## Filtering datasets

A lot of times we want to look at a small subset of a dataset as part of exploration.

We can use boolean indexing to filter the dataset into smaller subsets.

```python
df.region == 'Asia'
```

```python
df[df.region == 'Asia']
```

How about looking at only countries with GDP more than 10000.

```python
df[df.GDPperCapita > 20000]
```

How about countries in Asia where GDPperCapita is more than 20000?

```python
df[(df.region == 'Asia') & (df.GDPperCapita > 20000)]
```

We could also use the `query` method. The query method allows us to write simple expresisons like Python code.

```python
df.query("region == 'Asia' and GDPperCapita > 20000")
```

We can even refer local variables using `@` prefix.

```python
region = 'Asia'
GDP = 20000
```

```python
df.query("region == @region and GDPperCapita > @GDP")
```

## Group By

The Group-by is an operation that splits the dataframe into multiple parts, summarizes each part and combining the result.

For example, we may want to find sum of all values in a region.

```python
df.groupby("region").sum()
```

While the combined GDP in a region might make some sense, the other values doesn't really indicate anything in this case. 

Perhaps, computing _mean_ might be a good option here. 

```python
df.groupby("region").mean()
```

We can even do counts.

```python
df.groupby("region").count()
```

## Column Expressions

Pandas allows doing expressions on columns. These expressions will result in a new column (or a Series object).

For example, we may want to compute the ratio of lifeExpectency of Male vs. Female. 

```python
df['life_ratio'] = df.lifeMale / df.lifeFemale
```

```python
df.head()
```

### Problem: The Longest trains by duration

The trains dataset has two columns `duration_h` and `duration_m`. The first one indicates the number of hours and the second one indicates the number of minutes for the fractional hour. Combine these two columns to compute the total duration in minutes and use that to find the top 5 longest trains by duration.

```python
df_trains.head()
```

```python
df_trains['duration'] = df_trains.duration_h * 60 + df_trains.duration_m
```

```python
df_trains.sort_values("duration", ascending=False).head(10)
```
